```python
import requests
import urllib3
from bs4 import BeautifulSoup
import csv

```

```python

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36"
}

url = "https://www.sih.gov.in/sih2023-screening-final-result"
response = requests.get(url, headers=headers, verify=False)
soup = BeautifulSoup(response.content, "html.parser")


table = soup.find_all("table")[1]  
rows = table.find_all("tr")


headers = [header.text.strip() for header in rows[0].find_all("td")]


data = []
for row in rows[1:]:  
    cells = row.find_all("td")
    row_data = [cell.text.strip() for cell in cells]
    data.append(row_data)


with open("2023_screening_result.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(headers)  
    writer.writerows(data)     

print("Data saved to 2023_final_screening_result.csv")
```
